## *Data visualization*

### Analysis of Pregnancies

```python
fig1, ax1 = plt.subplots(1, 2, figsize=(20, 7))
fig2, ax2 = plt.subplots(figsize=(20, 7))

sns.histplot(data=data, x="Pregnancies", kde=True, ax=ax1[0],color = 'magenta')
sns.boxplot(data=data, x="Pregnancies", ax=ax1[1],color = 'aqua')

sns.violinplot(data=data, x="Pregnancies", ax=ax2,color = 'lime')

plt.show()
```

```python
print("Median of Pregnancies: ", data["Pregnancies"].median())
print("Maximum of Pregnancies: ", data["Pregnancies"].max())
```

## *Fill missing Values*

```python
# first copy the original dataframe into a new pandas dataframe
data_cp = data.copy()
# now fill the missing values with the median value of the particular columns
data_cp['Glucose'] = data['Glucose'].replace(0, data['Glucose'].median())
data_cp['BloodPressure'] = data['BloodPressure'].replace(0, data['BloodPressure'].median())
data_cp['SkinThickness'] = data['SkinThickness'].replace(0, data['SkinThickness'].median())
data_cp['Insulin'] = data['Insulin'].replace(0, data['Insulin'].median())
data_cp['BMI'] = data['BMI'].replace(0, data['BMI'].median())
```
